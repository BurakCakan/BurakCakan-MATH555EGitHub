## Question #1

There are four text files in the data folder

* Dicken's novel "Great Expectations" in English

* Calculate the character distributions, i.e. using the character counts, calculate the probability of each character appearing in the text.
* Find the set of characters common to all three texts.
* Using the common set and the KL-divergence, show that each language have different character distributions.
* Determine the language of the text file `unknown.txt` KL-divergence measure.

```python
import pandas as pd
from collections import Counter
from scipy import stats
from scipy.stats import entropy
import matplotlib.pyplot as plt

with open('/Users/burak.cakan/Desktop/HW2/ataturk_nutuk.txt', 'r') as file:
    data = file.read().replace('\n', '')
with open('/Users/burak.cakan/Desktop/HW2/dickens_great_expectations.txt', 'r') as file:
    data2 = file.read().replace('\n', '')
with open('/Users/burak.cakan/Desktop/HW2/flaubert_madame_bovary.txt', 'r') as file:
    data3 = file.read().replace('\n', '') 
Alphabet_TR=['a','b','c','ç','d','e','f','g','ğ','h','ı','i','j','k','l','m','n','o','ö','p','r','s','ş','t','u','ü','v','y','z']
Alphabet_EN=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
Alphabet_FR=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z','à','â','ç','é','ê','è','ë','î','ï','ô','ü','û','ù']


count_TR = {}
total_letters_TR=0
for i in range(0,29):
    count_TR[Alphabet_TR[i]] = data.lower().count(Alphabet_TR[i])
    total_letters_TR = total_letters_TR + data.lower().count(Alphabet_TR[i]) 
print("\nHow many times a distinct letter repeated in Nutuk:")
print(count_TR)
count_TR_p = {}
for i in range(0,29):
    count_TR_p[Alphabet_TR[i]] = count_TR[Alphabet_TR[i]]/total_letters_TR
print("\nProbability of each letter in Nutuk:")
print(count_TR_p)

total_letters_EN=0
count_EN = {}
for i in range(0,26):
    count_EN[Alphabet_EN[i]] = data2.lower().count(Alphabet_EN[i])
    total_letters_EN = total_letters_EN + data2.lower().count(Alphabet_EN[i]) 
print("\nHow many times a distinct letter repeated in Great Expectations:")
print(count_EN)
count_EN_p = {}
for i in range(0,26):
    count_EN_p[Alphabet_EN[i]] = count_EN[Alphabet_EN[i]]/total_letters_EN
print("\nProbability of each letter in Great Expectations:")
print(count_EN_p)

total_letters_FR=0
count_FR = {}
for i in range(0,39):
    count_FR[Alphabet_FR[i]] = data3.lower().count(Alphabet_FR[i])
    total_letters_FR = total_letters_FR + data3.lower().count(Alphabet_FR[i]) 
print("How many times a distinct letter repeated in Madame Bovary:")
print(count_FR)
count_FR_p = {}
for i in range(0,39):
    count_FR_p[Alphabet_FR[i]] = count_FR[Alphabet_FR[i]]/total_letters_FR
print("\nProbability of each letter in Madame Bovary:")
print(count_FR_p)

my_s1 = set(Alphabet_TR)
my_s2 = set(Alphabet_EN)
my_s3 = set(Alphabet_FR)
my_set1 = my_s1.intersection(my_s2)
output_set = my_set1.intersection(my_s3)
common_letters = list(output_set)
print("\nCommon letters in three alphabet are:")
print(common_letters)

#KL Divergence Calculation:

common_TR=[] 
for i in range(len(common_letters)):
    common_TR.append(count_TR_p[common_letters[i]])
common_FR=[]
for i in range(len(common_letters)):
    common_FR.append(count_FR_p[common_letters[i]])
common_EN=[]
for i in range(len(common_letters)):
    common_EN.append(count_EN_p[common_letters[i]])
    
print("\nKL Divergences of all pairs of probability distributions:")
print(entropy(common_TR,common_FR))
print(entropy(common_FR,common_TR))
print(entropy(common_FR,common_EN))
print(entropy(common_EN,common_FR))
print(entropy(common_EN,common_TR))
print(entropy(common_TR,common_EN))

#Bar chart of probability distributions:

plt.bar(common_letters, common_TR, align='center', alpha=0.5)
plt.ylabel('Probability Distribution')
plt.title('Prob. Dist. of Common Letters in TR')
plt.show()

plt.bar(common_letters, common_FR, align='center', alpha=0.5)
plt.ylabel('Probability Distribution')
plt.title('Prob. Dist. of Common Letters in FR')
plt.show()

plt.bar(common_letters, common_EN, align='center', alpha=0.5)
plt.ylabel('Probability Distribution')
plt.title('Prob. Dist. of Common Letters in EN')
plt.show()

# According to bar chart and KL Divergence, We can say that probability distributions are alike.

#Determining the languae of unknown text:
with open('/Users/burak.cakan/Desktop/HW2/unknown.txt', 'r') as file:
    data4 = file.read().replace('\n', '') 

count_UNK = {}
total_letters_UNK=0
for i in range(len(common_letters)):
    count_UNK[common_letters[i]] = data4.lower().count(common_letters[i])
    total_letters_UNK = total_letters_UNK + data4.lower().count(common_letters[i])
print("\nHow many times a common distinct letter repeated in Unknown text:")
print(count_UNK)
count_UNK_p = {}
for i in range(len(common_letters)):
    count_UNK_p[common_letters[i]] = count_UNK[common_letters[i]]/total_letters_UNK
print("\nProbability of each letter in Unknown text:")
print(count_UNK_p)

common_UNK=list(count_UNK_p.values())
#print(common_UNK)

print("\nKL Divergences of unkwnown text with the probability distributions of other languages:")
print(entropy(common_UNK,common_FR))
print(entropy(common_UNK,common_TR))
print(entropy(common_UNK,common_EN))

# We can clearly say that the probability distribution of unknown text is so similar to the probability distribution of English.
# Then, I verified that text is in English. So the assumption just above is right.
```

## Question #2

For this question consider the [Car Evaluation Data Set](https://archive.ics.uci.edu/ml/datasets/Car+Evaluation) from UCI. Here is the [direct link](https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data) to the dataset.

Make [contingency tables](https://en.wikipedia.org/wiki/Contingency_table#:~:text=In%20statistics%2C%20a%20contingency%20table,%2C%20engineering%2C%20and%20scientific%20research.) of the columns (using [`crosstab`](https://pandas.pydata.org/docs/reference/api/pandas.crosstab.html) function from [pandas](https://pandas.pydata.org)) and figure out which pairs of columns are dependent and independent. Explain your result using statistical tests.

```python
import pandas as pd 
import scipy.stats as stats

data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data',header=None) #, header = [0, 1], index_col = 0, engine = 'openpyxl'

data.columns = ['buying','maint','doors','persons','lug_boot','safety','class']
data_crosstab = pd.crosstab(data['buying'],data['class'],margins = False)
data_crosstab2 = pd.crosstab(data['maint'],data['class'],margins = False)
data_crosstab3 = pd.crosstab(data['doors'],data['class'],margins = False)
data_crosstab4 = pd.crosstab(data['persons'],data['class'],margins = False)
data_crosstab5 = pd.crosstab(data['lug_boot'],data['class'],margins = False)
data_crosstab6 = pd.crosstab(data['safety'],data['class'],margins = False)
#data_crosstab_all = pd.crosstab([data['buying'],data['safety']],data['class'])

print(data_crosstab)
print(data_crosstab2)
print(data_crosstab3)
print(data_crosstab4)
print(data_crosstab5)
print(data_crosstab6)
#print(data_crosstab_all)

#Becasue the target atttribute is class, I will test all categorical variables with class variable check whether they are dependent or not
#To do this, I will use chi-square test
# In the output of chi-square, second value is the p-value. And the hypothesis is as follows;
# H0=variables are independent; H1=variables are dependent; If p-value<0.05, Reject H0 (0.05 is a general threshold in literature)

buying_class=stats.chi2_contingency(data_crosstab) # Buying and class are dependent with p-value of 5.9280625992133936e-36
print("\nChi-square test for buying price and class:")
print(buying_class)
maint_class=stats.chi2_contingency(data_crosstab2) # Maint and class are dependent with p-value of 2.5476519845077733e-26
print("\nChi-square test for maintanence price and class:")
print(maint_class)
doors_class=stats.chi2_contingency(data_crosstab3) # Doors and class are independent with p-value of 0.32024215990030563
print("\nChi-square test for number of doors and class:")
print(doors_class)
persons_class=stats.chi2_contingency(data_crosstab4) # Persons and class are dependent with p-value of 4.039968047270742e-77
print("\nChi-square test for person capacity and class:")
print(persons_class)
lug_class=stats.chi2_contingency(data_crosstab5) # lug_boot and class are dependent with p-value of 1.0294402753134833e-09
print("\nChi-square test for luggage boot and class:")
print(lug_class)
safety_class=stats.chi2_contingency(data_crosstab6) # safety and class are dependent with p-value of 2.389155399044034e-100
print("\nChi-square test for luggage safety and class:")
print(safety_class)
```

## Question #4

For this question, use the [Iris Dataset](https://archive.ics.uci.edu/ml/datasets/iris) from UCI.  Here is the [direct link](https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data) to the dataset.

Your tasks are

* Form a [K-NN](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html) model for this dataset.
* Test your model on random samples of your data and calculate its accuracy.
* Repeat your calculation 100 times and give an interval of accuracy values leaving the best 2.5% and worst 2.5% accuracy values.
* Is there a better way of doing this without repeating the calculation 100 times? Explain.
* Find the best parameter $k$ for your dataset for the K-NN model.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',names=names) #, header = [0, 1], index_col = 0, engine = 'openpyxl'
print(data)

x = data.iloc[:, :-1].values
y = data.iloc[:, 4].values
#print(x) --> attributes
#print(y) --> target label

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
#print(len(x_train))
#print(len(x_test))

#to make normalization to prevent errors using distance measures;
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)

#Training:
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(x_train, y_train)

#Prediction:
y_pred = classifier.predict(x_test)

#these are the most common metrics to evaluate prediction:
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Accuracy= ",accuracy_score(y_test, y_pred))
#Results show that KNN predicted 30 data points with 100% precision and very high accuracy.

accuracy_interval=[]
for i in range(100):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20) #takes random sample from data
    scaler = StandardScaler()
    scaler.fit(x_train)
    x_train = scaler.transform(x_train)
    x_test = scaler.transform(x_test)
    classifier = KNeighborsClassifier(n_neighbors=5)
    classifier.fit(x_train, y_train)
    y_pred = classifier.predict(x_test)
    accuracy_interval.append(accuracy_score(y_test, y_pred))
min_acc=min(accuracy_interval)
max_acc=max(accuracy_interval)
sigma=((max_acc-min_acc)*(2.5))/100
print("\nAccuracy interval for 100 calculations is [",min_acc+sigma,",",max_acc-sigma,"]")
print("\n Instead of doing 100 calculations, the concept of k-fold cross validation is a better way to obtain a generalized accuracy level")

error = []
for i in range(1, 100):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    pred_i = knn.predict(x_test)
    error.append(np.mean(pred_i != y_test))
    
plt.figure(figsize=(12, 6))
plt.plot(range(1, 100), error, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
best_k=error.index(min(error))
print("\n Best k number for knn algorithm is ",best_k)
```
